```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and view

my_torch = my_torch.reshape(2,5)
my_torch
```

```python
my_torch2 = torch.arange(15)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(-1,5)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.view(2,5)
my_torch4
```

```python
#Difference between reshape and view, view does not create a copy, reshape does
```
