```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# Addition
tensor_a + tensor_b
```

```python
# addition longhand
torch.add(tensor_a, tensor_b)
```

```python
# Subtraction
tensor_b - tensor_a
```

```python
# sub Function
```

```python
torch.sub(tensor_b, tensor_a)
```

```python
# Multiplication
tensor_a * tensor_b
```

```python
torch.mul(tensor_a, tensor_b)
```

```python
# Division
tensor_b / tensor_a
```

```python
torch.div(tensor_b, tensor_a)
```

```python
#remainder modulus
tensor_b % tensor_a
```

```python
# Exponents
tensor_a ** tensor_b
```

```python
torch.pow(tensor_a, tensor_b)
```

```python
#Another way to longhand
tensor_a.add(tensor_b)
```

```python
tensor_a.sub(tensor_b)
```
